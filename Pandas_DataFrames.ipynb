DataFrame

It is a two dimensional heterogenous labeled data structure.
It is a collection of Series Data Structure.

```python
import pandas as pd

# Creating DataFrames using Series
purchase_1 = pd.Series({'Name': 'Chris',
                        'Item Purchased': 'Dog Food',
                        'Cost': 22.50})
purchase_2 = pd.Series({'Name': 'Kevyn',
                        'Item Purchased': 'Kitty Litter',
                        'Cost': 2.50})
purchase_3 = pd.Series({'Name': 'Vinod',
                        'Item Purchased': 'Bird Seed',
                        'Cost': 5.00})
# Column taken from key words and index values are passed
df1 = pd.DataFrame([purchase_1, purchase_2, purchase_3], index=['Store 1', 'Store 1', 'Store 2'])
df1.head()
```

```python
#Creating using list

l1 = ['Dog', 'bark', 'meat']
l2 = ['Cat', 'meow', 'meat']
l3= ['lion', 'roar', 'meat']

# Column value and index values are passed
df=pd.DataFrame([l1,l2,l3], index=['pet1','pet2','pet3'], columns=['animal', 'sound', 'food'])
print(df)
```

```python
df = pd.read_csv('olympics.csv', index_col = 0, skiprows =1)
df.head()
```

```python
df.columns #prints the column name.
```

```python
#https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.rename.html

#https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.replace.html

for colu in df.columns:
    if(colu[:2] == '01'):
        df.rename(columns={colu : 'Gold' + colu[4:] }, inplace =True)
    if(colu[:2]== '02'):
        df.rename(columns={colu: 'Silver' + colu[4:]}, inplace=True)
    if(colu[:2]== '03'):
        df.rename(columns={colu: 'Bronze' + colu[4:]}, inplace=True)
    if(colu[:1]=='№'):
        df.rename(columns={colu: '#' + colu[1:]}, inplace=True)
print(df.head())
```

Merging the dataframes

```python
df =  pd.DataFrame([{'Name': 'ramu', 'Item purchased' : 'Pencil', 'Cost' : 5.0 },
                    {'Name': 'sita', 'Item purchased': 'Pen', 'Cost': 10.0},
                    {'Name': 'Hanuman', 'Item purchased': 'Eraser', 'Cost': 7.0}],
                    index = ['Store1', 'Store2','Store2'])
df
```

```python
df['Date'] = ['Dec 1', 'Nov 20', 'Jan 2']
df
```

```python
df['Delivered'] = 'True'
df
```

```python
df ['Feedback'] = ['Good', None, 'bad']
df
```

```python
new_df = df.reset_index()
new_df
```

```python
new_df['Date'] =pd.Series( {0: 'Dec 1', 2: 'Jan 2'})
new_df
```

```python
df= pd.DataFrame([{'Name': 'ramu', 'Role': 'Student'},
                  {'Name': 'sita', 'Role': 'Doctor'}])
customer_df = new_df.set_index('Name')
customer_info = df.set_index('Name')
customer_info['Delivered'] =True
print(customer_df.head())
print(customer_info.head())
```

```python
#https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.merge.html

pd.merge(customer_df, customer_info, how = 'left', left_on='Name', right_on='Delivered')
```
