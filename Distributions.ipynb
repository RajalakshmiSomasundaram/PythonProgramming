```python
#Event e.g., tossing a coin, rolling a die etc
#Random variable - outcome of the event.
#Fair event - all outcomes should be equally distributed.
#Distribution - the shape that describes the probability of getting all possible random variables.
#NumPy and SciPy each have a number of different distributions built in for us to be able to sample from.

#Binomial distribution - there are only two possible outcomes 
# Also called as bernoulli distribution
#the probability of getting success is the same for each trail.
#the outcome distribution of a trial is independent and does not impact other trails.
# probability 0-1

#evenly weighed binomial distribution

np.random.binomial(10, 0.8 ) #number of trails, probability of getting success.
#o/p: no. of success
#np.random.binomial(1000, 0.5)/1000
```

```python
#unevenly weighed binomial distribution.

chance_of_getting_tornado = 0.01/100
np.random.binomial(1, chance_of_getting_tornado)
```

```python
#sampling the distribution

chance_of_tornado = 0.01

tornado_events = np.random.binomial(1, chance_of_tornado, 1000000)

tornado_events
```

```python

    
two_days_in_a_row = 0
for j in range(1,len(tornado_events)-1):
    if tornado_events[j]==1 and tornado_events[j-1]==1:
        two_days_in_a_row+=1

print('{} tornadoes back to back in {} years'.format(two_days_in_a_row, 1000000/365))

```

```python
#normal distribution or Gaussian distribution, Bell curve
#in this distribution, the probability of getting the number zero is high
#symmetric
# mean- for a given sample, finding the average of it
# expected value - what is the actual outcome (fair outcome)we are expecting from an event
#mean actually converges to the expected value given a sufficiently large sample set
#median - middle number of the data set
#mode - the number that occurred the most often
#variance - how far the distribution is spread out from the mean.
#standard deviation - how different is each value from the mean.

np.random.normal(-3,3)
distribution = np.random.normal(-3,3,size=10)
#np.std?
distribution
```

```python
#kurtosis-shape of the tail(measure of how heavily the tail is differ from normal distribution)
# -ve value - slightly more flat
# +ve value - more peaky
import scipy.stats as stats
stats.kurtosis(distribution)

#skew - push the peak of the distribution one way or the other
```

```python
#Chi Squared distribution
#Only one parameter called "degrees of freedom" - no. of samples taken from the normal distribution.
# less the degree of freedom - more skewed to the left.


chi_squared_df1 = np.random.chisquare(2, size=1000)
#stats.skew(chi_squared_df2)
chi_squared_df1
```

```python
chi_squared_df2 = np.random.chisquare(5, size=1000)
#stats.skew(chi_squared_df5)
chi_squared_df2
```

```python
#matplotlib - 2D ploting library
#pyplot - submodule of matplotlib used for statistical ploting

%matplotlib inline
import matplotlib
import matplotlib.pyplot as plt

output = plt.hist([chi_squared_df1,chi_squared_df2], bins=50, histtype='step', 
                  label=['2 degrees of freedom','5 degrees of freedom'])
plt.legend(loc='upper right')
```
